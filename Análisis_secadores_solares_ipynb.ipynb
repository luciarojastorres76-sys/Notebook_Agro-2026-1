```
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Datos
datos = {
    'humedad': [15.2, 18.5, 22.1, 14.1, 17.2, 21.8, 13.0, 16.8, 20.5, 12.0, 15.1, 18.9],
    'tratamiento': (['T1_Cabina']*3 + ['T2_Carpa']*3 + ['T3_Colector']*3 + ['T4_Chimenea']*3),
    'bloque': ['Dia_1', 'Dia_2', 'Dia_3'] * 4
}

df = pd.DataFrame(datos)

# Modelo DBCA
modelo = ols('humedad ~ tratamiento + bloque', data=df).fit()

# Tabla ANOVA
tabla_anova = sm.stats.anova_lm(modelo, typ=2)
print(tabla_anova)
```
