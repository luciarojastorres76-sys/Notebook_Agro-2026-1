## Ejemplos prácticos y sencillos

Aquí tienes tres aplicaciones directas que podrías usar hoy mismo:

### 1. Cálculo del Índice de Cosecha

Este código ayuda a determinar si un cultivo está listo para ser recolectado basándose en el peso del producto comercial frente al peso total.

```
# Datos de la muestra
peso_fruto = 450  # gramos
peso_planta_total = 1200  # gramos

# Cálculo del índice de cosecha
indice_cosecha = (peso_fruto / peso_planta_total) * 100

print(f"El índice de cosecha es del {indice_cosecha:.2f}%")
```

### 2. Conversión de Humedad (Base Húmeda a Base Seca)

En el secado de granos, es vital conocer la humedad exacta para evitar el crecimiento de hongos.

```
# Porcentaje de humedad en base húmeda
humedad_bh = 18  # ejemplo: 18%

# Fórmula para base seca
humedad_bs = humedad_bh / (100 - humedad_bh)

print(f"La humedad en base seca es: {humedad_bs:.4f} kg agua/kg sólido seco")
```

### 3. Alerta de Temperatura en Almacenamiento

Imagina un sensor conectado a Python que monitorea una bodega de frutas.

```
temperatura_actual = 28  # Grados Celsius

# Rango óptimo para manzanas, por ejemplo
temp_maxima = 25

if temperatura_actual > temp_maxima:
    print("¡ALERTA! Temperatura fuera de rango. Activando refrigeración...")
else:
    print("Temperatura estable. El producto está seguro.")
```
