**EJERCICIO 1: Capacidad de Silos**

📌 *Planteamiento del problema*

Se desea calcular cuántas toneladas de maíz puede almacenar un silo cilíndrico a partir de su radio y altura.

📌 *Requerimientos del problema*

Calcular el volumen del cilindro
Multiplicar por la densidad del maíz

📌 *Prueba del programa*

Radio = 3 m

Altura = 10 m

Resultado ≈ 212.06 toneladas

```python
import math

radio = float(input("Ingrese el radio del silo (m): "))
altura = float(input("Ingrese la altura del silo (m): "))

volumen = math.pi * (radio ** 2) * altura
toneladas = volumen * 0.75

print("La capacidad del silo es:", toneladas, "toneladas")
```

Explicación del código:
float(input()): convierte los datos a decimal
math.pi: constante π
radio ** 2: potencia (radio al cuadrado)
*: multiplicación
print(): muestra resultado

**Prueba de escritorio**
| Variable  | Valor  |
| --------- | ------ |
| radio     | 3      |
| altura    | 10     |
| volumen   | 282.74 |
| toneladas | 212.06 |


**EJERCICIO 2: Fertilizante**

📌 *Planteamiento*

Calcular cuánto fertilizante y agua se necesitan.

📌 *Requerimientos*

Multiplicar hectáreas × dosis
Aplicar relación 1:500

📌 *Prueba*

5 ha, 2 L/ha → 10 L producto, 5000 L agua

```python
hectareas = float(input("Ingrese las hectáreas: "))
dosis = float(input("Ingrese la dosis (L/ha): "))

producto = hectareas * dosis
agua = producto * 500

print("Producto necesario:", producto, "L")
print("Agua necesaria:", agua, "L")
```

Explicación:
Se usan float para datos decimales
Multiplicación directa
Relación proporcional

**Prueba de Escritorio**

| Variable  | Valor |
| --------- | ----- |
| hectareas | 5     |
| dosis     | 2     |
| producto  | 10    |
| agua      | 5000  |


**EJERCICIO 3: Rendimiento de pulpa**

📌 *Planteamiento*

Calcular la pulpa y su porcentaje.

📌 *Requerimientos*

Restar residuos
Calcular porcentaje

📌 *Prueba*

500 kg, 150 kg → 350 kg (70%)

```python
peso_total = float(input("Ingrese el peso total (kg): "))
residuos = float(input("Ingrese los residuos (kg): "))

pulpa = peso_total - residuos
rendimiento = (pulpa / peso_total) * 100

print("Pulpa obtenida:", pulpa, "kg")
print("Rendimiento:", rendimiento, "%")
```

Explicación:
- resta
/ división
* 100 porcentaje


**Prueba de escritorio**

| Variable | Valor |
| -------- | ----- |
| total    | 500   |
| residuos | 150   |
| pulpa    | 350   |
| %        | 70    |

**EJERCICIO 5: Viga de Acero**

📌 *Planteamiento*

Calcular el peso de una viga.

📌 *Requerimiento*

Volumen × densidad

📌 *Prueba*

Resultado: 7065 kg

```python
base = float(input("Ingrese la base (m): "))
altura = float(input("Ingrese la altura (m): "))
longitud = float(input("Ingrese la longitud (m): "))

peso = base * altura * longitud * 7850

print("El peso de la viga es:", peso, "kg")
```

Explicación:
Multiplicación de volumen
Uso de densidad constante

**Prueba de escritorio**
| Variable | Valor |
| -------- | ----- |
| base     | 0.3   |
| altura   | 0.5   |
| longitud | 6     |
| peso     | 7065  |


**EJERCICIO 6: Pendiente**

📌 *Planteamiento*

Calcular porcentaje de pendiente.

📌 *Requerimiento*

División × 100

📌 *Prueba*

0.6 / 5 → 12%

```python
altura = float(input("Ingrese la altura (m): "))
longitud = float(input("Ingrese la longitud (m): "))

pendiente = (altura / longitud) * 100

print("La pendiente es:", pendiente, "%")
```

Explicación:

División /
Conversión a porcentaje

**Prueba de Escritorio**
| Variable  | Valor |
| --------- | ----- |
| altura    | 0.6   |
| longitud  | 5     |
| pendiente | 12    |
